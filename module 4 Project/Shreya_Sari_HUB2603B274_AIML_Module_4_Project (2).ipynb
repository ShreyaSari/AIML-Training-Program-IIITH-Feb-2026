Why Gradient Descent is smoother?

Gradient Descent minimizes a continuous loss function, so:

It computes the average gradient over data
Updates are small and consistent
Noise from individual samples is reduced (especially in Batch / Mini-batch GD)

Perceptron, on the other hand:

1. Updates only when a sample is misclassified

2. Updates are discrete and abrupt

3. Highly sensitive to data ordering
